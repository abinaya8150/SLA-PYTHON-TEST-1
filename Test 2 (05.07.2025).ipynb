# 1) Merge two Python dictionaries into one
dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50}
Expected output
{'Ten': 10, 'Twenty': 20, 'Thirty': 30, 'Fourty': 40, 'Fifty': 50}


```python
dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50}
```

```python
merged_dict = dict1.copy()
```

```python
merged_dict.update(dict2)
print(merged_dict)
```

```python
merged_dict = {**dict1, **dict2}
print(merged_dict)
```

# 6) Write a python program to search a given number from a list
 use this list [4,5,6,2,3,9,1,4,5,6,3]

```python
numbers = [4,5,6,2,3,9,1,4,5,6,3]
search_num = int(input("Enter number to search: "))

if search_num in numbers:
    print(f"{search_num} found at index {numbers.index(search_num)}")
else:
    print(f"{search_num} not found in the list")
```

# 7) A palindrome number is a number that is same after reverse.


```python
num = int(input("Enter a number: "))

if str(num) == str(num)[::-1]:
    print(f"{num} is a palindrome number")
else:
    print(f"{num} is not a palindrome number")
```

# 8) Print multiplication table form 1 to 10 using for loops

```python
for i in range(1, 11):
    for j in range(1, 11):
        print(f"{i} x {j} = {i*j}", end='\t')
    print()
```

# 9)  The current population of a town is 10000. The population of the town is increasing at the rate
of 10% per year. You have to write a program to find out the population at the end of each of the last 10
years. For eg current population is 10000 so the output should be like this:
#10th year - 10000
#9th year - 9000
#8th year - 8100 and so on

```python
current_population = 10000

for year in range(10, 0, -1):
    print(f"{year}th year - {int(current_population)}")
    current_population /= 1.10
```

# 11) Print list in reverse order using a loop
use this list [1,2,3,4,5]


```python
numbers = [1, 2, 3, 4, 5]

for i in range(len(numbers)-1, -1, -1):
    print(numbers[i])
```
