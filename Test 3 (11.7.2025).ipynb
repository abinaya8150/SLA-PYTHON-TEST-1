# 1.#Accept a number user check whether it is prime or not (use for loop)

```python
num = int(input("Enter number: "))
if num > 1:
    
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

# 5) sorting with Lambda


```python
stu= [('alice',25), ('bob',20), ('charlie',30), ('allen',15)]
```

```python
name= sort(stu, key=lambda x: x[0])
print(name)
```

```python
age= sort(stu, key=lambda x: x[1])
print(age)
```

# 7) Write a recursive function to calculate the power of a number
#x ^ n.

```python
def power(x, n):
    if n == 0:
        return 1
    return x *power(x, n-1)

```

```python
print(power(2, 3))
```
