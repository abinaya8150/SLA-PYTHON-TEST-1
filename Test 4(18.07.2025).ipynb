```python
def sn(n):
    a=[1,1]
    for i in range(2,n+1):
        a.append(3*a[i-1]*a[i-2])
    return a
        
```

```python
print(sn(10))
```

```python
def sn(n):
    a=[1,1]
    for i in range(2,n+1):
        a.append(3*a[i-1]-a[i-2])
    return a
```

```python
print(sn(10))
```

```python
def rec(n):
    if n == 0:
        return 0
    return n + rec(n - 1)


print("recursive sm 0 to 10:", rec(10))
```
