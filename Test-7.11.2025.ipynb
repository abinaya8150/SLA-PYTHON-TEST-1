# 1. Print prime numbers from 2 to 30

```python
for num in range(2, 31):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

# 2.Transpose a matrix

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

transpose = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

print("Transpose of matrix:")
for row in transpose:
    print(row)
```

# 5. Extract numerical values from a list

```python
data = ['1', '2.0', 'hello', '9']
numbers = [x for x in data if x.replace('.', '', 1).isdigit()]
print(numbers)
```

```python
squares = [x**2 for x in range(1, 6)]
print(squares)
```

# 7. Demonstrate how pass, break, and continue work

```python
for i in range(5):
    if i == 2:
        continue   # Skips the current iteration
    elif i == 4:
        break      # Exits the loop
    elif i == 1:
        pass       # Does nothing
    print(i)
```

# 8. Extract consonants from the string 'ilovepython' using list comprehension

```python
string = 'ilovepython'
vowels = 'aeiou'
consonants = [ch for ch in string if ch not in vowels]
print(consonants)
```
