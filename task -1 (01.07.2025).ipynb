```python
# City-to-state mapping lists
tamil_nadu_cities = ["chennai", "trichy", "madurai", "coimbatore", "salem"]
karnataka_cities = ["bengaluru", "mysuru", "mangalore"]
kerala_cities = ["kochi", "thiruvananthapuram", "calicut"]
```

```python
# Input for city 2
city = input("Enter city  name: ")
if city in tamil_nadu_cities:
    print("City:{city}")
    print("State: Tamil Nadu")
elif city in karnataka_cities:
    print("City: {city}")
    print("State: Karnataka")
elif city in kerala_cities:
    print("City: {city}")
    print("State: Kerala")
else:
    print("City: {city}")
    print("State: Unknown")
print("Country: India\n")
```
